```python
from utils import kernels
model_Geyser = sequential_minimal_optimization.SMOModel(X_Geyser, Y_Geyser, 10.0, kernels.GaussianKernel(5))
model_Chips = sequential_minimal_optimization.SMOModel(X_Chips, Y_Chips, 50.0, kernels.GaussianKernel(1))
model_Geyser.train()
model_Chips.train()
```
